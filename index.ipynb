## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There are three types of relationships between entities in a SQL database.  They are one-to-one, one-to-many, and many-to-many.  A one-to-one relationship means that each key relates to only one other key in a separate table.  A one-to-many relationship is a little more complex.  This involves one key in a table referencing multiple keys in a seperate table.  For example, one table may refer to an office using an office code, but in a seperate table there can be several employees who each are assigned to this office code.  The single office code key refers to many employees in a seperate table.  Finally there is the many-to-many relationship.  In this relationship there are several keys between tables that reference each other.  For example, imagine a table that lists offices in several cities.  In a separate table there is a list of customers in each of those cities.  This is a many-to-many relationship as the city column can reference each office and each customer.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There are 4 types of JOINs with SQL.  The first is an inner join.  This type will return any keys that are found in both tables.  Then there is the left join, which takes the entirety of the left table and adds any matching keys from the right table.  A default join command is a left join.  Next is the right join, which is the inverse of the left join.  The entire right table is returned with any matching keys from the left table.  Finally, there is the outer join which is the inverse of the inner join.  It will return any keys that are unique to one of the tables.


## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Aggregate functions are a convenient way of calculating sums, averages and counts of information stored within tables.  Group by statements allow us to take information stored within a table and sort it into like categories.  By combining the two of these together, we are able to compute the aggregate functions of groups, rather than the aggregate functions of the entire table.  For instance, we can compute the average count of employees who work in each city, rather than just the count of all of the employees.
